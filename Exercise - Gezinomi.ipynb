```python
def categorize_booking_days(days):
    if days <= 7:
        return "Last Minuters"
    elif 7 < days <= 30:
        return "Potential Planners"
    elif 30 < days <= 90:
        return "Planners"
    else:
        return "Early Bookers"

df['EB_Score'] = df['SaleCheckInDayDiff'].apply(categorize_booking_days)
```

```python
kirilim1 = df.groupby(["SaleCityName","ConceptName","EB_Score"])["Price"].agg(["mean","count"])
```

```python
agg_df = df.groupby(["SaleCityName", "ConceptName", "Seasons"])["Price"].mean().sort_values(ascending=False)
```

```python
agg_df
```

```python
agg_df = pd.DataFrame(agg_df)
agg_df
```

```python
agg_df.reset_index(inplace = True)
```

```python
agg_df["sales_level_based"] = agg_df["SaleCityName"].str.upper() + "_" + agg_df["ConceptName"].str.upper()  + "_" + agg_df["Seasons"].str.upper() 
```

```python
agg_df.head()
del agg_df["SaleCityName"]
del agg_df["ConceptName"]
del agg_df["Seasons"]
```

```python
agg_df["Segment"] = pd.qcut(agg_df["Price"], 4, labels=["D", "C", "B", "A"])
```

```python
agg_df.sample(3)
```

```python
agg_df.groupby("Segment")["Price"].agg(["mean", "max", "sum"])
```
